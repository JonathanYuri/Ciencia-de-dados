Ler todas tabelas de abonos

```python
def fAbono(data, media_AB, numAB):
  for i in data:
    ler = "ABONOP_"
    ler = ler + i + '.csv'

    dados = pd.read_csv(ler, encoding='latin-1', sep=';')

    # NUMERO DE PESSOAS ABONADAS
    numAB.append(len(dados.index))

    # MEDIA DE VALOR GASTO COM ABONO
    media = dados['Val'].mean()

    if (pd.isna(media) == True):
      media_AB.append(0)
    else:
      media_AB.append(round(media, 0))
```

Ler todas tabelas de aposentados

```python
def fAposentados(data, media_AP, media_APV, media_API, numAP, numAPV, numAPI):
  for i in data:
    ler = "APOSENTADOS_"
    ler = ler + i + '.csv'

    dados = pd.read_csv(ler, encoding='latin-1', sep=';')

    # NUMERO DE PESSOAS APOSENTADAS
    numAP.append(len(dados.index))

    # MEDIA DE VALOR GASTO COM APOSENTADOS
    media = dados['VALOR MENSAL DA APOSENTADORIA'].mean()

    if (pd.isna(media) == True):
      media_AP.append(0)
    else:
      media_AP.append(round(media, 0))

    dfb = dados[dados['DESCRIÌO DO TIPO DE APOSENTADORIA'] == 'VOLUNTARIA                 ']

    # MEDIA DE VALOR GASTO COM APOSENTADOS VOLUNTARIOS
    media = dfb['VALOR MENSAL DA APOSENTADORIA'].mean()

    # NUMERO DE APOSENTADOS VOLUNTARIOS
    numAPV.append(len(dfb.index))
  
    if (pd.isna(media) == True):
      media_APV.append(0)
    else:
      media_APV.append(round(media, 0))

    dfb = dados[dados['DESCRIÌO DO TIPO DE APOSENTADORIA'] != 'VOLUNTARIA                 ']

    # MEDIA DE VALOR GASTO COM APOSENTADOS NÃO VOLUNTARIOS
    media = dfb['VALOR MENSAL DA APOSENTADORIA'].mean()
    
    # NUMERO DE APOSENTADOS NAO VOLUNTARIAS
    numAPI.append(len(dfb.index))
    
    if (pd.isna(media) == True):
      media_API.append(0)
    else:
      media_API.append(round(media, 0))
```
